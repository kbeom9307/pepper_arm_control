## keyboard event & move with input coordinate

```python
"""
Move work_pose first, then move 5 centimeters toward left (y axis positive side).
"""

import time
import numpy as np
import naoqi as n
import sys 
import pepper_kinematics as pk
import tkinter as tk
import Euler_Angles 

nao_ip = "192.168.0.39"
port = 9559

m = n.ALProxy("ALMotion", nao_ip, port)
dcm = n.ALProxy("DCM", nao_ip, port)
m.setAngles(pk.left_arm_tags, pk.left_arm_work_pose, 0.05) # friction speed

time.sleep(1.0)

current_angles = []
for i in range(len(pk.left_arm_tags)): # ALMotion.getanlges arg - str 로 변환
    current_angles.append(m.getAngles(pk.left_arm_tags[i],0))
    
current_angles = [i[0] for i in current_angles] # list 벗겨내고
current_position, current_orientation = pk.left_arm_get_position(current_angles)

target_position = current_position
target_orientation = np.array(Euler_Angles.rot2euler(current_orientation))


root = tk.Tk()
root.title("motion GUI")
root.minsize(width=300, height= 300)

tk.Label(root, text = "Choose Mode.").pack(side="top")

target_ori_euler = np.array(Euler_Angles.rot2euler(current_orientation))

def left(event):
    print("move left : 1cm")
    target_position[1] = target_position[1] + 0.01 # 1 cm move left
    global target_angles
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def right(event):
    print("move right : 1cm")
    target_position[1] = target_position[1] - 0.01 # 1 cm move right
    global target_angles 
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def foward(event):
    print("move foward : 1cm")
    target_position[0] = target_position[0] + 0.01 # 1 cm move foward
    global target_angles
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def backward(event):
    print("move backward : 1cm")
    target_position[0] = target_position[0] - 0.01 # 1 cm move backward
    global target_angles
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def up(event):
    print("move up : 1cm")
    target_position[2] = target_position[2] + 0.01 # 1 cm move up
    global target_angles
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def down(event):
    print("move down : 1cm")
    target_position[2] = target_position[2] - 0.01 # 1 cm move down
    global target_angles
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def roll_p(event):
    print("roll up")
    target_orientation[0] = target_orientation[0] + 0.01
    global target_anlges
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)

def pitch_p(event):
    print("pitch up")
    target_orientation[1] = target_orientation[1] + 0.01
    global target_anlges
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def yaw_p(event):
    print("yaw up")
    target_orientation[2] = target_orientation[2] + 0.01
    global target_anlges
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)

def roll_m(event):
    print("roll down")
    target_orientation[0] = target_orientation[0] - 0.01
    global target_anlges
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)

def pitch_m(event):
    print("pitch down")
    target_orientation[1] = target_orientation[1] - 0.01
    global target_anlges
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)

def yaw_m(event):
    print("yaw down")
    target_orientation[2] = target_orientation[2] - 0.01
    global target_anlges
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
def quit(event):
    print("총 이동 거리는 : ? , 좌표는 : ? ")
    global root
    root.destroy()
    
def callback():
    frame1.focus_set() # 키보드에 focus
    frame1.bind("<Left>", left)
    frame1.bind("<Right>", right)
    frame1.bind("<Up>", foward)
    frame1.bind("<Down>", backward)
    frame1.bind("<Shift-Up>", up)
    frame1.bind("<Shift-Down>", down)
    frame1.bind("<Escape>", quit)
    frame1.bind("<q>",roll_p)
    frame1.bind("<e>",roll_m)
    frame1.bind("<w>",pitch_p)
    frame1.bind("<s>",pitch_m)
    frame1.bind("<a>",yaw_p)
    frame1.bind("<d>",yaw_m)
    frame1.pack()
    
def callback2():
    
    lbl = tk.Label(frame2, text = "Coordinate")
    lbl.pack()
    
    xlbl = tk.Label(frame2, text = "X" )
    xlbl.pack()
    global x
    x = tk.Entry(frame2, width = 10)
    x.pack()
    x.focus_set() # 포커스 설정
    
    ylbl = tk.Label(frame2, text = "Y")
    ylbl.pack()
    global y
    y = tk.Entry(frame2, width = 10)
    y.pack()
    
    zlbl = tk.Label(frame2, text = "Z")
    zlbl.pack()
    global z
    z = tk.Entry(frame2, width = 10)
    z.pack()  
  
    lbl2 = tk.Label(frame2, text = " Orientation RPY (unit : deg) ")
    lbl2.pack()
  
    rlbl = tk.Label(frame2, text = "Roll")
    rlbl.pack()
    global roll
    roll = tk.Entry(frame2, width =10)
    roll.pack()
    
    plbl = tk.Label(frame2, text = "Pitch")
    plbl.pack()
    global pitch
    pitch = tk.Entry(frame2, width = 10)
    pitch.pack()
    
    ylbl = tk.Label(frame2, text = "Yaw")
    ylbl.pack()
    global yaw
    yaw = tk.Entry(frame2, width = 10)
    yaw.pack()

    btnmove = tk.Button(frame2, text="Move", command = movecoordinate)
    btnmove.pack(side = "bottom")

    btnsave = tk.Button(frame2, text="Save", command = saveVar)
    btnsave.pack(side = "bottom")
    
def callback3():

    if left.get() == 1:
        handName = "LHand"
        while 1: 
            dcm.set(["Device/SubDeviceList/LHand/Position/Actuator/Value", "Merge", [[0.0, dcm.getTime(10000)]] ])        
            time.sleep(5)
            break
    
    if right.get() == 1:
        handName = "RHand"
        while True:
            dcm.set(["Device/SubDeviceList/RHand/Position/Actuator/Value", "Merge", [[0.0, dcm.getTime(10000)]] ])
            time.sleep(5)
            break
            
def callback4():
    
    if left.get() == 1:
        handName = "LHand"
        while 1:
            dcm.set(["Device/SubDeviceList/LHand/Position/Actuator/Value", "Merge", [[1.0, dcm.getTime(10000)]] ])        
            time.sleep(5)
            break
            
    if right.get() == 1:
        handName = "RHand"
        while True:
            dcm.set(["Device/SubDeviceList/RHand/Position/Actuator/Value", "Merge", [[1.0, dcm.getTime(10000)]] ])
            time.sleep(5)
            break
    
def saveVar():
    
    global move_position_x, move_position_y, move_position_z 
    move_position_x = float(x.get())/100
    move_position_y = float(y.get())/100
    move_position_z = float(z.get())/100
    
    global move_ori_roll, move_ori_pitch, move_ori_yaw
    move_ori_roll  = float(roll.get())*(math.pi/180)     # degree to radian
    move_ori_pitch = float(pitch.get())*(math.pi/180)
    move_ori_yaw   = float(yaw.get())*(math.pi/180)
                
    print("Save arg X",move_position_x," Y",move_position_y," Z",move_position_z)
    print("Save deg R",move_ori_roll," P", move_ori_pitch," Y", move_ori_yaw)
    
def movecoordinate():
    
    target_position[0] = move_position_x
    target_position[1] = move_position_y
    target_position[2] = move_position_z
    
    target_orientation[0] = move_ori_roll
    target_orientation[1] = move_ori_pitch
    target_orientation[2] = move_ori_yaw
    
    
    global target_angles
    target_angles = pk.left_arm_set_position(current_angles, target_position, target_orientation)
    m.setAngles(pk.left_arm_tags, list(target_angles), 0.05)
    
    
frame1 = tk.Frame(root, width=300, height=300)
frame1.pack(side="top", expand = "true", anchor = "n")

frame2 = tk.Frame(root, width=300, height=300)
frame2.pack(side="top", expand = "true" , anchor ="n")

frame3 = tk.Frame(root, width=300, height=300)
frame3.pack(side="bottom", expand = "true", anchor ="s")

btn=tk.Button(frame1, text="control", command = callback)
btn.pack(side ="top")

btn2=tk.Button(frame2, text="place", command = callback2)
btn2.pack(side ="top")

left = tk.IntVar()
check1 = tk.Checkbutton(frame3, text="left", variable = left)
check1.grid(column=0, row=0)

right = tk.IntVar()
check2 = tk.Checkbutton(frame3, text="right", variable = right)
check2.grid(column=0, row=1)                        

btn3 = tk.Button(frame3, text="open", command = callback4)
btn3.grid(column=0, row=2)

btn4 = tk.Button(frame3, text="grip", command = callback3)
btn4.grid(column=1, row=2)

root.mainloop()
```
